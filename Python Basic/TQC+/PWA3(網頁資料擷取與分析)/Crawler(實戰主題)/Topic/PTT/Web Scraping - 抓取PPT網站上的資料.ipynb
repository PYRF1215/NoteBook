##### ． 查看網頁伺服器回應請求的狀態
####  ．抓取資料

```python
Cookie_URL = 'https://www.ptt.cc/ask/over18?from=%2Fbbs%2FGossiping%2Findex.html'

# Session() 將我們送出 requests 所收到的 cookies 全部儲存起來，且於下一次發送請求時送出對應的參數。
r = requests.Session( ) 

# 設定 Cookies 內容
load = { 'from':'/bbs/Gossiping/index.html', 'yes' : 'yes' }

# post( ) 函數送出 post 請求 
r1 = r.post( Cookie_URL, data = load )

# 查看網頁伺服器回應 post( ) 請求的狀態
print( '網頁伺服器回應 post( ) 請求的狀態：' )
try:
    if r1.status_code == 200 : print( '請求成功！' )
    r1.raise_for_status( )    # 取得請求錯誤的詳細資訊！
except  requests.exceptions.RequestException as ex1 : print( 'HTTP 請求錯誤 : ' + str( ex1 ) )
except  requests.exceptions.HTTPError as ex2 : print( 'HTTP 回應錯誤 : ' + str( ex2 ) )
except  requests.exceptions.ConnectionError as ex4 : print( '網路連線錯誤 : ' + str( ex4 ) )
print( '---------------------' '\n'  )
print(  '\n' '網頁原始碼： ' '\n'  )
print( r1.text )
```

```python
sp = BeautifulSoup( r1.text, "html.parser" ) # 使用 BeautifulSoup，將從網頁抓取下來的資料轉為 html.parser。
print( sp )
```

```python
# 範例：取出文章的標題

# 取 HTML 標中的 <div class="title">
#                        .........
#                        </div> 中的 <a> 標籤存入 tag_title
title = sp.select( 'div.title a' ) # 使用 select( ) 函數，取出 HTML 網頁中的資料。
print( title )
```

```python
for i in title:
    print( i["href"], i.text ) # 印出網址及文章標題
```
